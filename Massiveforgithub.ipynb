```python
#Here I will make a list of the tracks for a given playlist
#This is essential for deleting the tracks already in a playlist because
#My goal is to update these playlists. 

def build_track_list(playlist_id):
    offset = 0
    track_ids=[]
    while True:
        response = sp.playlist_tracks(playlist_id,
                                      offset=offset,
                                      fields='items.track.id,total')
        items=response['items']
        for track in items:
            if track['track']:
                track_ids.append(track['track']['id'])

        offset = offset + len(response['items'])
        print(offset, "/", response['total'])

        if len(response['items']) == 0:
            break
    return track_ids
```

```python
#Here I will remove the tracks from a given playlist. This can use the list created in the 
# build_track_list function. I should have used this function at the top.
def track_remover(playlist_id,track_ids):
    for chunk in chunker(track_ids,100):
        if token:
            sp = spotipy.Spotify(auth=token)
            sp.trace = False
            results = sp.user_playlist_remove_all_occurrences_of_tracks(
                username, playlist_id, chunk)
    #         pprint.pprint(results)
        else:
            print("Can't get token for", username)
```
